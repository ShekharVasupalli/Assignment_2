Q1) Generate a list of 100 integers containing values between 90 to 130 and store it in the variable `int_list`. After generating the list, find the        following:

  

  (i) Write a Python function to calculate the mean of a given list of numbers.Create a function to find the median of a list of numbers.

  

  (ii) Develop a program to compute the mode of a list of integers.

  

  (iii) Implement a function to calculate the weighted mean of a list of values and their corresponding weights.

  

  (iv) Write a Python function to find the geometric mean of a list of positive numbers.

  

  (v) Create a program to calculate the harmonic mean of a list of values.

  

  (vi) Build a function to determine the midrange of a list of numbers (average of the minimum and maximum).

  

  (vii) Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of outliers.

```python
import random
from collections import Counter
from typing import List, Tuple

# Generate a list of 100 integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]

# (i) Mean and Median functions
def calculate_mean(numbers: List[int]) -> float:
    return sum(numbers) / len(numbers)

def calculate_median(numbers: List[int]) -> float:
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    mid = n // 2
    if n % 2 == 0:
        return (sorted_numbers[mid - 1] + sorted_numbers[mid]) / 2
    else:
        return sorted_numbers[mid]

# (ii) Mode function
def calculate_mode(numbers: List[int]) -> int:
    count = Counter(numbers)
    mode_data = count.most_common(1)
    return mode_data[0][0] if mode_data else None

# (iii) Weighted Mean function
def calculate_weighted_mean(values: List[int], weights: List[int]) -> float:
    weighted_sum = sum(value * weight for value, weight in zip(values, weights))
    total_weight = sum(weights)
    return weighted_sum / total_weight

# (iv) Geometric Mean function
def calculate_geometric_mean(numbers: List[float]) -> float:
    product = 1
    for number in numbers:
        product *= number
    return product ** (1 / len(numbers))

# (v) Harmonic Mean function
def calculate_harmonic_mean(numbers: List[float]) -> float:
    reciprocal_sum = sum(1 / number for number in numbers)
    return len(numbers) / reciprocal_sum

# (vi) Midrange function
def calculate_midrange(numbers: List[int]) -> float:
    return (min(numbers) + max(numbers)) / 2

# (vii) Trimmed Mean function
def calculate_trimmed_mean(numbers: List[int], trim_percent: float) -> float:
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    trim_count = int(n * trim_percent / 100)
    trimmed_numbers = sorted_numbers[trim_count:n - trim_count]
    return sum(trimmed_numbers) / len(trimmed_numbers)

# Testing the functions with the generated list
mean = calculate_mean(int_list)
median = calculate_median(int_list)
mode = calculate_mode(int_list)
weighted_mean = calculate_weighted_mean(int_list, [1] * len(int_list))  # Equal weights as an example
geometric_mean = calculate_geometric_mean(int_list)
harmonic_mean = calculate_harmonic_mean(int_list)
midrange = calculate_midrange(int_list)
trimmed_mean = calculate_trimmed_mean(int_list, 10)  # Example with 10% trimming

mean, median, mode, weighted_mean, geometric_mean, harmonic_mean, midrange, trimmed_mean
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a list of 500 integers between 200 and 300
int_list2 = [random.randint(200, 300) for _ in range(500)]

# (i) Visualization of Frequency & Gaussian distribution, Frequency KDE plot, and Gaussian & KDE
def plot_visualizations(data: List[int]):
    plt.figure(figsize=(18, 10))

    # Frequency & Gaussian distribution
    plt.subplot(1, 3, 1)
    sns.histplot(data, bins=15, kde=False, color='skyblue', label='Frequency')
    sns.kdeplot(data, color='red', label='Gaussian')
    plt.title('Frequency & Gaussian Distribution')
    plt.legend()

    # Frequency smoothened KDE plot
    plt.subplot(1, 3, 2)
    sns.histplot(data, bins=15, kde=True, color='lightgreen', label='Frequency + KDE')
    plt.title('Frequency with Smoothened KDE Plot')
    plt.legend()

    # Gaussian distribution & smoothened KDE plot
    plt.subplot(1, 3, 3)
    sns.kdeplot(data, color='purple', label='Gaussian')
    sns.kdeplot(data, color='orange', bw_adjust=1.5, label='Smoothened KDE')
    plt.title('Gaussian Distribution & Smoothened KDE Plot')
    plt.legend()

    plt.tight_layout()
    plt.show()

# (ii) Range function
def calculate_range(numbers: List[int]) -> int:
    return max(numbers) - min(numbers)

# (iii) Variance and Standard Deviation
def calculate_variance(numbers: List[int]) -> float:
    mean_value = calculate_mean(numbers)
    return sum((x - mean_value) ** 2 for x in numbers) / len(numbers)

def calculate_standard_deviation(numbers: List[int]) -> float:
    return calculate_variance(numbers) ** 0.5

# (iv) Interquartile Range (IQR)
def calculate_iqr(numbers: List[int]) -> float:
    q1 = np.percentile(numbers, 25)
    q3 = np.percentile(numbers, 75)
    return q3 - q1

# (v) Coefficient of Variation (CV)
def calculate_coefficient_of_variation(numbers: List[int]) -> float:
    mean_value = calculate_mean(numbers)
    std_dev = calculate_standard_deviation(numbers)
    return (std_dev / mean_value) * 100

# (vi) Mean Absolute Deviation (MAD)
def calculate_mad(numbers: List[int]) -> float:
    mean_value = calculate_mean(numbers)
    return sum(abs(x - mean_value) for x in numbers) / len(numbers)

# (vii) Quartile Deviation
def calculate_quartile_deviation(numbers: List[int]) -> float:
    return calculate_iqr(numbers) / 2

# (viii) Range-based Coefficient of Dispersion
def calculate_range_coefficient_of_dispersion(numbers: List[int]) -> float:
    range_value = calculate_range(numbers)
    min_value = min(numbers)
    max_value = max(numbers)
    return range_value / (min_value + max_value)

# Generate visualizations
plot_visualizations(int_list2)

# Testing the functions with the generated list
range_value = calculate_range(int_list2)
variance = calculate_variance(int_list2)
standard_deviation = calculate_standard_deviation(int_list2)
iqr = calculate_iqr(int_list2)
coefficient_of_variation = calculate_coefficient_of_variation(int_list2)
mad = calculate_mad(int_list2)
quartile_deviation = calculate_quartile_deviation(int_list2)
range_coefficient_of_dispersion = calculate_range_coefficient_of_dispersion(int_list2)

range_value, variance, standard_deviation, iqr, coefficient_of_variation, mad, quartile_deviation, range_coefficient_of_dispersion
```

Q3) Write a Python class representing a discrete random variable with methods to calculate its expected 
value and variance.

```python
from typing import List

class DiscreteRandomVariable:
    def __init__(self, values: List[int], probabilities: List[float]):
        
        if len(values) != len(probabilities):
            raise ValueError("Values and probabilities lists must have the same length.")
        if abs(sum(probabilities) - 1) > 1e-6:
            raise ValueError("Probabilities must sum to 1.")
        
        self.values = values
        self.probabilities = probabilities

    def expected_value(self) -> float:
        """
        Calculate the expected value (mean) of the discrete random variable.
        
        :return: Expected value of the random variable.
        """
        return sum(value * prob for value, prob in zip(self.values, self.probabilities))

    def variance(self) -> float:
        """
        Calculate the variance of the discrete random variable.
        
        :return: Variance of the random variable.
        """
        mean = self.expected_value()
        return sum(((value - mean) ** 2) * prob for value, prob in zip(self.values, self.probabilities))

# Example usage:
values = [1, 2, 3, 4, 5]
probabilities = [0.1, 0.2, 0.3, 0.2, 0.2]
discrete_rv = DiscreteRandomVariable(values, probabilities)

expected_value = discrete_rv.expected_value()
variance = discrete_rv.variance()

print("Expected Value:", expected_value)
print("Variance:", variance)
```

Q5) Create a Python function to generate random samples from a given probability distribution (e.g., binomial, Poisson) and calculate their mean and variance.

```python
import numpy as np
from typing import List

def generate_samples(distribution: str, params: dict, sample_size: int = 1000) -> dict:
    
    if distribution == 'binomial':
        # Generate samples from a binomial distribution
        n = params.get('n')
        p = params.get('p')
        if n is None or p is None:
            raise ValueError("For binomial distribution, parameters 'n' and 'p' are required.")
        samples = np.random.binomial(n, p, sample_size)
    elif distribution == 'poisson':
        # Generate samples from a Poisson distribution
        lam = params.get('lambda')
        if lam is None:
            raise ValueError("For Poisson distribution, parameter 'lambda' is required.")
        samples = np.random.poisson(lam, sample_size)
    else:
        raise ValueError("Supported distributions are 'binomial' and 'poisson'.")

    # Calculate mean and variance of the samples
    sample_mean = np.mean(samples)
    sample_variance = np.var(samples)

    return {
        'samples': samples,
        'mean': sample_mean,
        'variance': sample_variance
    }

# Generate binomial samples with n=10, p=0.5
binomial_result = generate_samples('binomial', {'n': 10, 'p': 0.5})
print("Binomial Distribution (n=10, p=0.5):")
print("Sample Mean:", binomial_result['mean'])
print("Sample Variance:", binomial_result['variance'])

# Generate Poisson samples with lambda=4
poisson_result = generate_samples('poisson', {'lambda': 4})
print("\nPoisson Distribution (lambda=4):")
print("Sample Mean:", poisson_result['mean'])
print("Sample Variance:", poisson_result['variance'])
```

Q6) Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute the mean, variance, and standard deviation of the samples.

```python
import numpy as np

def generate_gaussian_samples(mean: float, std_dev: float, sample_size: int = 1000) -> dict:
    """
    Generate random samples from a Gaussian (normal) distribution and calculate their mean, variance, and standard deviation.
    
    :param mean: The mean () of the Gaussian distribution.
    :param std_dev: The standard deviation () of the Gaussian distribution.
    :param sample_size: Number of random samples to generate.
    :return: A dictionary containing the samples, mean, variance, and standard deviation.
    """
    # Generate samples from a Gaussian distribution
    samples = np.random.normal(loc=mean, scale=std_dev, size=sample_size)

    # Calculate mean, variance, and standard deviation of the samples
    sample_mean = np.mean(samples)
    sample_variance = np.var(samples)
    sample_std_dev = np.std(samples)

    return {
        'samples': samples,
        'mean': sample_mean,
        'variance': sample_variance,
        'standard_deviation': sample_std_dev
    }

# Generate Gaussian samples with mean=50 and standard deviation=10
gaussian_result = generate_gaussian_samples(mean=50, std_dev=10)

print("Gaussian Distribution (mean=50, std_dev=10):")
print("Sample Mean:", gaussian_result['mean'])
print("Sample Variance:", gaussian_result['variance'])
print("Sample Standard Deviation:", gaussian_result['standard_deviation'])
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, pearsonr

# Load the tips dataset
tips = sns.load_dataset("tips")

# (i) Function to calculate skewness of a column
def calculate_skewness(column: pd.Series) -> float:
    return skew(column)

# (ii) Function to determine skewness type
def skewness_type(column: pd.Series) -> str:
    skewness_value = calculate_skewness(column)
    if skewness_value > 0:
        return "Positive Skewness"
    elif skewness_value < 0:
        return "Negative Skewness"
    else:
        return "Approximately Symmetric"

# (iii) Function to calculate covariance between two columns
def calculate_covariance(column1: pd.Series, column2: pd.Series) -> float:
    return np.cov(column1, column2)[0, 1]

# (iv) Function to calculate Pearson correlation coefficient between two columns
def calculate_pearson_correlation(column1: pd.Series, column2: pd.Series) -> float:
    return pearsonr(column1, column2)[0]

# (v) Script to visualize correlation using scatter plot
def plot_scatter_with_correlation(column1: pd.Series, column2: pd.Series, x_label: str, y_label: str) -> None:
    plt.figure(figsize=(8, 5))
    plt.scatter(column1, column2, alpha=0.5, color="b")
    plt.title(f"Scatter Plot of {x_label} vs {y_label}")
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid(True)
    plt.show()

# Analysis of 'total_bill' and 'tip' columns
total_bill = tips['total_bill']
tip = tips['tip']

# (i) Skewness
total_bill_skewness = calculate_skewness(total_bill)
tip_skewness = calculate_skewness(tip)
print(f"Skewness of total_bill: {total_bill_skewness}")
print(f"Skewness of tip: {tip_skewness}")

# (ii) Skewness type
print(f"Total bill skewness type: {skewness_type(total_bill)}")
print(f"Tip skewness type: {skewness_type(tip)}")

# (iii) Covariance
covariance = calculate_covariance(total_bill, tip)
print(f"Covariance between total_bill and tip: {covariance}")

# (iv) Pearson correlation coefficient
pearson_corr = calculate_pearson_correlation(total_bill, tip)
print(f"Pearson correlation coefficient between total_bill and tip: {pearson_corr}")

# (v) Scatter plot to visualize correlation
plot_scatter_with_correlation(total_bill, tip, "Total Bill", "Tip")
```

Q8) Write a Python function to calculate the probability density function (PDF) of a continuous random variable for a given normal distribution.

```python
import math

def normal_pdf(x: float, mean: float, std_dev: float) -> float:
    
    exponent = -((x - mean) ** 2) / (2 * std_dev ** 2)
    pdf_value = (1 / (std_dev * math.sqrt(2 * math.pi))) * math.exp(exponent)
    return pdf_value

# Example usage:
mean = 0  # Mean of the distribution
std_dev = 1  # Standard deviation of the distribution
x = 1  # Point at which to calculate the PDF

pdf_value = normal_pdf(x, mean, std_dev)
print(f"The PDF value at x = {x} for N({mean}, {std_dev}^2): {pdf_value}")
```

Q9) Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.

```python
import math

def exponential_cdf(x: float, rate: float) -> float:

    if x < 0:
        return 0.0  # CDF is 0 for x < 0
    if rate <= 0:
        raise ValueError("Rate parameter must be positive.")

    return 1 - math.exp(-rate * x)

# Example usage:
rate = 0.5  # Rate parameter ()
x_values = [0, 1, 2, 3, 4, 5]  # Points to evaluate the CDF

# Calculate and print the CDF for each x value
for x in x_values:
    cdf_value = exponential_cdf(x, rate)
    print(f"CDF at x = {x} for Exp(={rate}): {cdf_value}")
```

Q10) Write a Python function to calculate the probability mass function (PMF) of Poisson distribution.

```python
import math

def poisson_pmf(k: int, rate: float) -> float:
  
    if k < 0:
        raise ValueError("Number of occurrences (k) must be a non-negative integer.")
    if rate <= 0:
        raise ValueError("Rate parameter must be positive.")

    return (math.exp(-rate) * (rate ** k)) / math.factorial(k)

# Example usage:
rate = 4.0  # Rate parameter ()
k_values = [0, 1, 2, 3, 4, 5]  # Number of occurrences to evaluate the PMF

# Calculate and print the PMF for each k value
for k in k_values:
    pmf_value = poisson_pmf(k, rate)
    print(f"PMF at k = {k} for Poisson(={rate}): {pmf_value}")
```

```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest


old_layout = np.array([1] * 50 + [0] * 950)
new_layout = np.array([1] * 70 + [0] * 930)

# Calculate the number of successes (purchases) and number of trials (visitors)
successes = [np.sum(old_layout), np.sum(new_layout)]
trials = [len(old_layout), len(new_layout)]

# Perform the Z-test
z_stat, p_value = proportions_ztest(successes, trials)

print(f"Z-statistic: {z_stat}")
print(f"P-value: {p_value}")

# Determine the significance level
alpha = 0.05  # Common significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the conversion rates.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the conversion rates.")
```

Q15) A company is testing two different website layouts to see which one leads to higher click-through rates. 
Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees o 
freedom, and p-valu

Use the following data:

'''
python

layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]

layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]
'''e.

```python
import numpy as np
from scipy import stats

def ab_test(layout_a_clicks, layout_b_clicks):
    # Convert lists to numpy arrays for calculations
    layout_a = np.array(layout_a_clicks)
    layout_b = np.array(layout_b_clicks)
    
    mean_a = np.mean(layout_a)
    mean_b = np.mean(layout_b)
    
    std_a = np.std(layout_a, ddof=1)  # Sample standard deviation
    std_b = np.std(layout_b, ddof=1)  # Sample standard deviation
    
    # Calculate the t-statistic
    t_stat, p_value = stats.ttest_ind(layout_a, layout_b, equal_var=False)
    
    # Calculate degrees of freedom
    n_a = len(layout_a)
    n_b = len(layout_b)
    dof = (std_a**2 / n_a + std_b**2 / n_b) ** 2 / \
          ((std_a**2 / n_a)**2 / (n_a - 1) + (std_b**2 / n_b)**2 / (n_b - 1))

    print(f"Mean Clicks Layout A: {mean_a:.2f}")
    print(f"Mean Clicks Layout B: {mean_b:.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"Degrees of Freedom: {dof:.2f}")
    print(f"P-value: {p_value:.4f}")

    # Interpretation
    alpha = 0.05  # Common significance level
    if p_value < alpha:
        print("Reject the null hypothesis: There is a significant difference in click-through rates between the two layouts.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference in click-through rates between the two layouts.")

# Data provided
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Perform the A/B test
ab_test(layout_a_clicks, layout_b_clicks)
```

```python
import numpy as np
from scipy import stats

def analyze_cholesterol(existing_drug_levels, new_drug_levels):
    # Convert lists to numpy arrays for calculations
    existing_drug = np.array(existing_drug_levels)
    new_drug = np.array(new_drug_levels)
    
    # Calculate means and standard deviations
    mean_existing = np.mean(existing_drug)
    mean_new = np.mean(new_drug)
    
    std_existing = np.std(existing_drug, ddof=1)  # Sample standard deviation
    std_new = np.std(new_drug, ddof=1)  # Sample standard deviation
    
    # Perform the t-test (independent two-sample)
    t_stat, p_value = stats.ttest_ind(new_drug, existing_drug, equal_var=False)  # Welch's t-test
    
    print(f"Mean Cholesterol Levels (Existing Drug): {mean_existing:.2f}")
    print(f"Mean Cholesterol Levels (New Drug): {mean_new:.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpretation
    alpha = 0.05  # Common significance level
    if p_value < alpha:
        print("Reject the null hypothesis: The new drug is significantly more effective in reducing cholesterol levels.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference in effectiveness between the new drug and the existing drug.")

# Data provided
existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]
new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]

# Analyze cholesterol levels
analyze_cholesterol(existing_drug_levels, new_drug_levels)
```

Q21) A political analyst wants to determine if there is a significant association between age groups and voter preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify them into different age groups and candidate preferences. Perform a Chi-Square test to determine if there is a significant association between age groups and voter preferences.

Use the below code to generate data:

'''
python

np.random.seed(0)

age_groups = np.random.choice(['18 30','31 50','51+','51+'], size=30)

voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)
'''

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Generate data
np.random.seed(0)
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=500)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500)

# Create a DataFrame
data = pd.DataFrame({'Age Group': age_groups, 'Voter Preference': voter_preferences})

# Create a contingency table
contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])

# Perform the Chi-Square test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Print the results
print("Contingency Table:")
print(contingency_table)
print("\nChi-Square Statistic:", chi2_stat)
print("P-value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Frequencies Table:")
print(expected)

# Interpretation
alpha = 0.05  # Common significance level
if p_value < alpha:
    print("\nReject the null hypothesis: There is a significant association between age groups and voter preferences.")
else:
    print("\nFail to reject the null hypothesis: There is no significant association between age groups and voter preferences.")
```
